```python
!pip install transformers arabert datasets -q

import pandas as pd
import numpy as np
import random
import torch
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from arabert.preprocess import ArabertPreprocessor
from transformers import (
    AutoTokenizer, AutoModelForSequenceClassification,
    Trainer, TrainingArguments, DataCollatorWithPadding,
    EarlyStoppingCallback
)
from datasets import Dataset

# =========================================================
# 1️⃣ Load and Prepare Dataset
# =========================================================
df = pd.read_csv("/content/Tweets - VT  (1).csv")
df = df.rename(columns={"Content": "text", "Target": "label"})
df = df[df["label"].isin(["yes", "no"])].dropna(subset=["text", "label"])
df = shuffle(df, random_state=42).reset_index(drop=True)
```

```python
# Split data (train / val / test)
train_df, temp_df = train_test_split(df, test_size=0.3, stratify=df["label"], random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.5, stratify=temp_df["label"], random_state=42)
```

```python
# 2️⃣ Simple Arabic Augmentation (no Farasa)
# =========================================================
def simple_augment(text, p_drop=0.1, p_shuffle=0.1):
    """Light Arabic augmentation: drop or shuffle small portions of text"""
    words = text.split()
    if len(words) < 4:
        return text
    # Drop some words
    words = [w for w in words if random.random() > p_drop]
    # Random shuffle small subset
    if random.random() < p_shuffle:
        random.shuffle(words)
    return " ".join(words)

minority_df = train_df[train_df["label"] == "yes"]
majority_df = train_df[train_df["label"] == "no"]

augmented_texts = []
while len(augmented_texts) + len(minority_df) < len(majority_df):
    for text in minority_df["text"]:
        augmented_texts.append(simple_augment(text))
        if len(augmented_texts) + len(minority_df) >= len(majority_df):
            break

augmented_df = pd.DataFrame({"text": augmented_texts, "label": ["yes"] * len(augmented_texts)})
balanced_train_df = pd.concat([majority_df, minority_df, augmented_df], ignore_index=True)
balanced_train_df = shuffle(balanced_train_df, random_state=42).reset_index(drop=True)

print("Balanced training set size:", len(balanced_train_df))
balanced_train_df["label"].value_counts().plot(kind="bar", title="Balanced Class Distribution")
```

```python
# 3️⃣ Tokenization & Preprocessing
# =========================================================
label_map = {"no": 0, "yes": 1}
for d in [balanced_train_df, val_df, test_df]:
    d["label"] = d["label"].map(label_map)

arabert_prep = ArabertPreprocessor("aubmindlab/bert-base-arabertv02-twitter")
balanced_train_df["text"] = balanced_train_df["text"].apply(arabert_prep.preprocess)
val_df["text"] = val_df["text"].apply(arabert_prep.preprocess)
test_df["text"] = test_df["text"].apply(arabert_prep.preprocess)

hf_train = Dataset.from_pandas(balanced_train_df)
hf_val = Dataset.from_pandas(val_df)
hf_test = Dataset.from_pandas(test_df)

tokenizer = AutoTokenizer.from_pretrained("aubmindlab/bert-base-arabertv02-twitter")

def tokenize(batch):
    return tokenizer(batch["text"], truncation=True, max_length=192)

encoded_train = hf_train.map(tokenize, batched=True, remove_columns=["text"])
encoded_val = hf_val.map(tokenize, batched=True, remove_columns=["text"])
encoded_test = hf_test.map(tokenize, batched=True, remove_columns=["text"])

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
# 4️⃣ Lightweight Model Setup
# =========================================================
model = AutoModelForSequenceClassification.from_pretrained(
    "aubmindlab/bert-base-arabertv02-twitter", num_labels=2
)

model.config.hidden_dropout_prob = 0.4
model.config.attention_probs_dropout_prob = 0.4
model.gradient_checkpointing_enable()

# Class weights for mild imbalance
class_counts = balanced_train_df["label"].value_counts().to_dict()
weights = torch.tensor([1.0 / class_counts[0], 1.0 / class_counts[1]])
device = "cuda" if torch.cuda.is_available() else "cpu"
weights = weights.to(device)
loss_fct = torch.nn.CrossEntropyLoss(weight=weights)

def custom_compute_loss(model, inputs, return_outputs=False):
    labels = inputs.pop("labels")
    outputs = model(**inputs)
    logits = outputs.logits
    loss = loss_fct(logits, labels)
    return (loss, outputs) if return_outputs else loss
```

```python
# 5️⃣ Training Configuration
# =========================================================
training_args = TrainingArguments(
    output_dir="./results_light",
    eval_strategy="epoch",
    save_strategy="epoch",
    learning_rate=3e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.05,
    load_best_model_at_end=True,
    metric_for_best_model="f1",
    greater_is_better=True,
    logging_steps=50,
    report_to=[],
    seed=42,
    fp16=True,
)

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=1)
    acc = accuracy_score(labels, preds)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average="binary")
    return {"accuracy": acc, "precision": precision, "recall": recall, "f1": f1}

# Custom Trainer to use the custom loss function
class CustomTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False, num_items_in_batch=None): # Added num_items_in_batch
        labels = inputs.pop("labels")
        outputs = model(**inputs)
        logits = outputs.logits
        loss = loss_fct(logits, labels)
        return (loss, outputs) if return_outputs else loss

trainer = CustomTrainer(
    model=model,
    args=training_args,
    train_dataset=encoded_train,
    eval_dataset=encoded_val,
    tokenizer=tokenizer,
    data_collator=data_collator,
    compute_metrics=compute_metrics,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=1)],
)

# =========================================================
# 6️⃣ Train & Evaluate
# =========================================================
trainer.train()

metrics = trainer.evaluate(encoded_test)
print("Final Test Metrics:", metrics)

# Confusion Matrix
predictions = trainer.predict(encoded_test)
y_pred = np.argmax(predictions.predictions, axis=1)
y_true = encoded_test["label"]
cm = confusion_matrix(y_true, y_pred)
ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Not Viral", "Viral"]).plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()

# Save model
trainer.save_model("./viraLens_light_best")
tokenizer.save_pretrained("./viraLens_light_best")
print("✅ Light model saved successfully as viraLens_light_best/")
```
